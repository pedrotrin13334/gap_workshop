```gap
max_number := 1000;;
primes_counted := 0;;
acc_primes_list := [];;
for number in [1..max_number] do
    if IsPrime(number) then
        primes_counted := primes_counted +1;;
    fi;
    Add(acc_primes_list, primes_counted);;
od;

title1 := Concatenation("Quantidade de n√∫meros primos at√© ordem ", String(max_number));
Plot([1..max_number], acc_primes_list, rec(title := title1));

x_log_list := List([1..max_number], x->Float(x)/Log(Float(x)));;

division_list := List([1..max_number], x-> Float(acc_primes_list[x]/x_log_list[x]));;

Plot([1..max_number], division_list, rec(title:= "Rela√ß√£o assint√≥tica entre x/log(x) e a quantidade de n√∫meros primos"));
```

```gap
max_order := 200;;
groups_tested := 0;;
abelian_counted := 0;;
percent_list := [];;
for order in [1..max_order] do
    group_list := AllSmallGroups(order);;
    groups_tested := groups_tested + Length(group_list);;
    abelian_groups := AllSmallGroups(order, x->IsAbelian(x));;
    abelian_counted := abelian_counted + Length(abelian_groups);;
    Add(percent_list, Float(abelian_counted/groups_tested));;
od;

title1:= Concatenation("Raz√£o de n√∫meros abelianos at√© ordem ", String(max_order));;
Plot([1..max_order], percent_list, rec(title:= title1));
```
